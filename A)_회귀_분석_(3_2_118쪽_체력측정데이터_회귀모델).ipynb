# 다중 선형 회귀(125쪽)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('fitness data.csv', encoding='cp949')
df.head()
```

```python
fig, ax=plt.subplots(1,3, figsize=(14,5))
sns.histplot(df['반복점프'], ax=ax[0])
sns.histplot(df['왕복오래달리기'], ax=ax[1])
sns.histplot(df['제자리멀리뛰기'], ax=ax[2])
plt.show()
```

```python
X=df[['반복점프','왕복오래달리기']]
y=df[['제자리멀리뛰기']]
```

```python
model2=LinearRegression()
```

```python
model2.fit(X,y)
print(model2.coef_, model2.intercept_)
```

#회귀 모델의 평가와 비교(127쪽)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model3=LinearRegression()
```

```python
model3.fit(X_train, y_train)
```

```python
pred=model3.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, pred)
```

```python
r2_score(y_train, model3.predict(X_train))
```

```python
X=df[:10][['반복점프','왕복오래달리기']]
y=df[:10][['제자리멀리뛰기']]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=0)
from sklearn.linear_model import LinearRegression
model4=LinearRegression()
model4.fit(X_train,y_train)
from sklearn.metrics import r2_score
print(r2_score(y_test,model4.predict(X_test)), r2_score(y_train,
                                                        model4.predict(X_train)))
```
