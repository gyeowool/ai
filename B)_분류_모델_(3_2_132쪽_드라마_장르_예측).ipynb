```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib
df = pd.read_csv('drama genre.csv', index_col=0)
df.head()
```

```python
fig, ax = plt.subplots(7, 2, figsize=(20,15))
for i in range(14):
  sns.barplot(x=df.genre, y=df.iloc[:,i+1], estimator=np.mean, ax=ax[i%7,i//7])
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='genre', data=df)
```

```python
top=df[df.genre.isin(['일일연속극', '미니시리즈', '주말연속극'])]
```

```python
X=top.iloc[:,1:]
y=top['genre']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
LR.predict([X_test.iloc[0]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, LR.predict(X_test))
```

```python
from sklearn.metrics import precision_score, recall_score
print(precision_score(y_test,LR.predict(X_test), average='macro'))
print(recall_score(y_test,LR.predict(X_test), average='macro'))
```
