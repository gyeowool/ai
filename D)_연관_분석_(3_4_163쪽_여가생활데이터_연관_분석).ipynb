```python
import pandas as pd
df=pd.read_csv('여가생활.csv')
df.head()
```

```python
df.rename(columns={'ONE_YEAR_WITHIN_EXPR_LSR_CLTUR_VALUE':
                   '여가생활'}, inplace=True)
for i in range(df.shape[0]):
  df['여가생활'][i]=df['여가생활'][i].replace(' ','').split(',')
```

```python
from mlxtend.preprocessing import TransactionEncoder
te=TransactionEncoder()
te_ary=te.fit(df['여가생활']).transform(df['여가생활'])
items=pd.DataFrame(te_ary, columns=te.columns_)
items.head()
```

```python
top=items.sum().sort_values(ascending=False).head(10)
top10=items[top.index]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frequent_itemsets = apriori(top10, min_support=0.1, use_colnames=True)
```

```python
association_rules(frequent_itemsets, metric="confidence",
                  min_threshold=0.6).sort_values(by=['confidence','lift'], ascending=False)
```

```python
rules=association_rules(frequent_itemsets, metric="confidence",min_threshold=0.6)
inc_movie=[]
for i in range(rules.shape[0]):
  if '영화관관람' in list(rules.loc[i,'consequents']):
    inc_movie.append(i)
rules.iloc[inc_movie,:]
```
